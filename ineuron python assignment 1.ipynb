```python
'''
2. Create the below pattern using nested for loop in Python.

'''

for i in range(0,9):
    if i<=4:
        for j in range(0,i+1):
            print("*",end=" ")
        print("\n")
    else:
        for j in range(i,9):
            print("*",end=" ")
        print("\n")
        
```
